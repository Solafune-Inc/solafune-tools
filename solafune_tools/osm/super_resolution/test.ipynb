```python
from inference import Model
import time
import tifffile
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Time the inference on a sample normal image from warming up the model to running the model
start_time = time.time()
model = Model()
image = tifffile.TiffFile("sample_input/small.tif").asarray()
result = model.generate(image)
end_time = time.time()
print(f"Inference time: {end_time - start_time:.2f} seconds")
# Display the result side by side with the original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('on')
plt.subplot(1, 2, 2)
plt.title("Super-Resolved Image")
plt.imshow(result, cmap='gray')
plt.axis('on')
plt.show()
```
